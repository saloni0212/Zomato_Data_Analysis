```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
df=df.drop(['url','address','phone','menu_item','listed_in(type)','listed_in(city)'],axis=1)
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
feature_na = [i for i in df.columns if df[i].isnull().sum() > 0]
feature_na
```

These are the columns which contains NULL values

```python
for i in feature_na:
    print(f"{i} has {np.round((df[i].isnull().sum()/len(df[i])*100),2)}% null values")
```

```python
df.rate.unique()
```

```python
df.dropna(subset=['rate'], axis=0, inplace=True)
```

```python
def split(x):
    return x.split("/")[0].strip() # splitting with "/" with split() and removing trailing and leading whitespaces using strip() func
```

```python
df['rate'] = df['rate'].apply(split)
```

```python
df['rate']
```

```python
df.rate.unique()
```

```python
df['rate'].replace(['NEW','-'],0,inplace=True) # replacing 'NEW' and '-' with 0
```

```python
df.info()
```

```python
df.rate = df['rate'].astype(float) # changing type from 'object' to 'float'
```

```python
df.info()
```

## Calculate average rating for each restaurant

```python
rating = pd.pivot_table(df, index='name', values='rate')
rating
```

```python
rating = rating.sort_values(['rate'], ascending=False)
rating[0:15] # top 15 restaurants
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=rating[0:20].rate, y=rating[0:20].index, orient="h")
plt.show() # visualization of top 15 restaurants
```

### Distribution of ratings

```python
sns.set_style('whitegrid')
sns.distplot(df['rate'])
plt.show()
```

This is not a normal distribution as we can see a spike in '0's i.e the new restaurants

```python
from scipy.stats import normaltest

DataToTest = df['rate']
stat, p = normaltest(DataToTest)
print("stat=%0.2f, p=%0.30f" % (stat, p))

if(p>0.05):
    print("Normal distribution")
else:
    print("Not a normal distribution")
```

### Let's analyse the highly rated restaurants and worst restaurants based on number of ratings given

```python
df.head(3)
```

```python
sns.distplot(df['votes'])

plt.show()
```

```python
voting_rating = df.groupby('name')[['votes']].mean().sort_values('votes', ascending=False)
voting_rating
```

```python
voting_rating['name'] = voting_rating.index
```

```python
voting_rating = voting_rating.reset_index(drop = True)
voting_rating.head()
```

```python
voting_rating = pd.merge(voting_rating, df[['rate','name']])
voting_rating
```

```python
vote_top = voting_rating[voting_rating['votes'] >5000]
vote_top.head()
```

```python
vote_top = vote_top.groupby('name')[['rate']].mean().sort_values('rate', ascending=False)
vote_top
```

### Highest Rated based on most number of votes

```python
plt.figure(figsize=(20,8), dpi=100)
sns.barplot(x=vote_top['rate'], y=vote_top.index, orient="h")

plt.show()
```

```python
bad_count = voting_rating[(voting_rating['rate']<3) & (voting_rating['rate'] > 0)]
bad_count
```

```python
bad_restaurants = bad_count[bad_count['votes'] > 500].groupby('name')[['rate']].mean()
bad_restaurants
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x=bad_restaurants['rate'], y=bad_restaurants.index)
plt.show()
```
